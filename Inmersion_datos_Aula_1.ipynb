**Desafíos de esta aula**

1. Promedio del área de todos los inmuebles en los barrios en el dataset. El top 10.

Calculamos el promedio de las areas por barrio

````python
promedio_barrio = inmuebles[["Barrio", "Area"]].groupby("Barrio").mean()
promedio_barrio
````

Ordenamos de promedio mayor al menor y mostramos el top 10

````python
top_barrios_area = promedio_barrio.sort_values(["Area"], ascending=False).head(10)
top_barrios_area
````

Graficamos

````python
top_barrios_area.plot.bar()
````

Podemos ver los inmuebles con mayor área de Modelia para confirmar la diferencia

````python
inmuebles_Modelia = (inmuebles.Barrio == "Modelia")
Modelia = inmuebles[inmuebles_Modelia]
Modelia.sort_values(["Area"], ascending=False).head(10)
````
